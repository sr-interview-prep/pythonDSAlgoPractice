```python
s = "leetcode"
```

```python
def firstUniqChar(s: str) -> int:
    hashMap=set()
    for i in range(len(s)):
        if s[i] not in hashMap:
            hashMap.add(s[i])
        else:
            hashMap.remove(s[i])
    for i in range(len(s)):
        if s[i] not in hashMap:
            return i
    return -1
        
```

```python
firstUniqChar(s)
```
